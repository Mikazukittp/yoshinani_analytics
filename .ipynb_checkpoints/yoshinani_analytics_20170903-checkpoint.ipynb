# LINEユーザのアクティブ率は高いのか分析

```python
query = """
SELECT
    U.id AS account_id,
    MAX(P.created_at) AS lastest_payment_date, 
    MIN(U.created_at) AS register_date 
FROM users AS U 
LEFT OUTER JOIN payments AS P 
    ON U.id = P.paid_user_id 
GROUP BY account_id
"""
```

```python
users_payment = pd.read_sql_query(query, connection)
```

```python
# 一度でも投稿したことあるユーザの割合(全体)
users_payment.query('lastest_payment_date not in [@pd.NaT]')["account_id"].count() / users_payment["account_id"].count()
```

```python
# LINEユーザだけで見てみると
```

```python
query = """
SELECT
    U.id AS account_id,
    MAX(P.created_at) AS lastest_payment_date, 
    MIN(U.created_at) AS register_date 
FROM users AS U 
INNER JOIN oauth_registrations AS OU 
    ON U.id = OU.user_id
LEFT OUTER JOIN payments AS P 
    ON U.id = P.paid_user_id
GROUP BY account_id
"""
```

```python
line_users_payment = pd.read_sql_query(query, connection)
```

```python
line_users_payment.query('lastest_payment_date not in [@pd.NaT]')["account_id"].count() / line_users_payment["account_id"].count()
```

** → 逆に悪化してる。流入元によってアクティブ率はそんなに変わらない。**

- ** グループの属性文字列で切ってくの難しい。。旅行って行っても地名入れてる人もいるし。やっぱデータ設計大事 **

```python
query = """
SELECT
    a.size AS '人数',
    count(a.payment_id) AS 'レコード数'
FROM(
    SELECT
        count(id) AS size,
        payment_id
    FROM
        participants
    GROUP BY
        payment_id
) AS a
GROUP BY
    a.size
ORDER BY
    a.size
"""
```

# 1支払いにおけるユーザ割合
- 圧倒的にユーザ二人での利用が多い

```python
payments = pd.read_sql_query(query, connection)
```

```python
payments['割合'] = 100 * payments['レコード数'] / payments['レコード数'].sum()
payments
```
